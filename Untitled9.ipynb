```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import time
from matplotlib.animation import FuncAnimation
import os

import matplotlib.animation as animation

q = 0.6
p = 1 - q
n = 10000
frame = 100
dict = st.uniform.rvs(0,100,n)
la = np.zeros(n + 1)
m = np.zeros(n + 1)
mult = n // frame
plt.hist(dict)


def mu(i, a, b):
    return 1 / (a + b * i)


for i, el in enumerate(dict):
    la[i + 1] = la[i] + mu(i, 0.9, 0.001) * (q if (el >= la[i]) else -p)
    m[i + 1] = m[i] + mu(i, 0.5, 0.5) * (el - m[i])

numDataPoints = n

path_script = os.getcwd()
x = []
lax = []
mx = []


def animate_func(num):
    x.append(num * mult)
    lax.append(la[num * mult])
    mx.append(m[num * mult])
    ax.clear()
    ax.set_xlim(0, n)
    ax.set_ylim(dict.min(), dict.max())
    ax.scatter(x[:num], lax[:num], c='blue')
    ax.scatter(x[:num], mx[:num], c='red')
    ax.plot([0, n], [dict.mean(), dict.mean()], color='black')
    ax.plot([0, n], [np.quantile(dict, q), np.quantile(dict, q)], color='yellow')


fig, ax = plt.subplots(figsize=(12, 10))

line_ani = animation.FuncAnimation(fig, animate_func, interval=1 / 10,
                                   frames=100)
line_ani.save(os.getcwd() + '/grid.gif', fps=20, writer='pillow')
plt.show()
```

```python
import pygame
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *
import random
import numpy as np
from PIL import Image
from PIL import ImageOps
import matplotlib.pyplot as plt
def gausis():
    return sum([random.random() for i in range(12)]) - 6


def rndP(sigma, korr, size,f):
    n=gausis()
    mas = [f]
    for i in range(1,size):
        mas.append(korr * f + sigma * (1 - korr ** 2) ** (1 / 2) * n)
        f = mas[i]
        n = gausis()
    return mas


def hasbula(sigm, korr1, korr2, display):
    mas = np.zeros((display[0], display[1]))
    f=sigm*gausis()
    mas[0][0]=f
    mas[0, 1:] = rndP(sigm, korr1, display[0]-1,f)
    mas[1:,0] = rndP(sigm, korr2, display[1]-1,f)
    for i in range(1,display[0]):
        for j in range(1,display[1]):
            mas[i][j] = (korr1 * mas[i][j - 1] + korr2 * mas[i - 1][j] - korr1 * korr2 * mas[i - 1][j - 1] + sigm * (
                    ((1 - korr1 ** 2) * (1 - korr2 ** 2)) ** (1 / 2)) * gausis())

    glBegin(GL_POINTS)
    k = -display[0]
    mas=np.array(mas)
    for i in range(display[0]):
        k += 2
        l = display[1]
        for j in range(display[1]):
            glColor3d(mas[i][j],mas[i][j],mas[i][j])
            glVertex2d(k / display [0], l / display [1])
            l -= 2

    glEnd()
    plt.imshow(mas, vmin = -3, vmax = 3, cmap = 'gray')
    print(mas)


def main():
    pygame.init()
    display = (50, 50)
    pygame.display.set_mode(display, DOUBLEBUF | OPENGL)
    hasbula(1, 0.5, -1,np.array(display))
    glPixelStorei(GL_PACK_ALIGNMENT, 1)
    img_buf = glReadPixelsub(0, 0, display[1], display[0], GL_RGB, GL_UNSIGNED_BYTE)
    img = Image.frombytes("RGB", (display[1], display [0]), img_buf)
    img.save('glutout.png', 'PNG')
    pygame.display.flip()

main()
```

```python
from __future__ import division
from linear_algebra import squared_distance, vector_mean, distance
import math, random
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

class KMeans:
    """performs k-means clustering"""

    def __init__(self, k):
        self.k = k          # number of clusters
        self.means = None   # means of clusters

    def classify(self, input):
        """return the index of the cluster closest to the input"""
        return min(range(self.k),
                   key=lambda i: squared_distance(input, self.means[i]))

    def train(self, inputs):

        self.means = random.sample(inputs, self.k)
        assignments = None
        print(self.k)
        while True:
            # Find new assignments
            new_assignments = list(map(self.classify, inputs))

            # If no assignments have changed, we're done.
            if assignments == new_assignments:
                return

            # Otherwise keep the new assignments,
            assignments = new_assignments

            for i in range(self.k):
                i_points = [p for p, a in zip(inputs, assignments) if a == i]
                # avoid divide-by-zero if i_points is empty
                if i_points:
                    self.means[i] = vector_mean(i_points)


def squared_clustering_errors(inputs, k):
    """finds the total squared error from k-means clustering the inputs"""
    clusterer = KMeans(k)
    clusterer.train(inputs)
    means = clusterer.means
    assignments = map(clusterer.classify, inputs)

    return sum(squared_distance(input,means[cluster])
               for input, cluster in zip(inputs, assignments))



if __name__ == "__main__":

    data = pd.read_csv('austin_crime.csv',header=0, names = ["clearance_date","clearance_code","clearance_status","description_code","description","district_code","district","latitude","longitude","timestamp","time"])
    data = data.drop(["clearance_date","clearance_code","clearance_status","description_code","description","district","latitude","longitude","timestamp"],axis=1)
    x = data[:-1] #tipo de crime e status de clarificao

    #inputs = np.array(x)
    inputs = np.array(x)
    print (inputs)
    print (len(inputs))

    x = [row[0] for row in inputs]
    print (x)
    y = [row[1] for row in inputs]
    print (y)
    plt.scatter(x, y)
    plt.title("Tempo por distrito para encerramento de crimes")
    plt.xlabel("distrito")
    plt.ylabel("tempo")
    plt.show()


    #ks = range(1, len(inputs) + 1)
    ks = range(1, 20)
    errors = [squared_clustering_errors(list(inputs), k) for k in ks]
    plt.plot(ks, errors)
    plt.xticks(ks)
    plt.xlabel("k")
    plt.ylabel("total squared error")
    plt.show()


    random.seed(0)
    clusterer = KMeans(2)

    clusterer.train(inputs)
    print ("2-means:")
    print (clusterer.means)

    plt.plot(x, y, 'b*')
    plt.title("distribuicao pontos")
    plt.xlabel("x")
    plt.ylabel("y")
    clusterX = [row[0] for row in clusterer.means]
    clusterY = [row[1] for row in clusterer.means]
    plt.plot(clusterX, clusterY, 'rs')
    plt.show()



    random.seed(0) # so you get the same results as me
    clusterer = KMeans(3)
    clusterer.train(inputs)
    print ("3-means:")
    print (clusterer.means)
    print ()

    plt.plot(x, y, 'b*')
    plt.title("distribuicao pontos")
    plt.xlabel("x")
    plt.ylabel("y")
    clusterX = [row[0] for row in clusterer.means]
    clusterY = [row[1] for row in clusterer.means]
    plt.plot(clusterX, clusterY, 'rs')
    plt.show()
```
